```python
import yfinance as yf

import matplotlib.pyplot as plt

import pandas as pd

ticker = input('Enter Stock Name: ')

ticker = ticker.upper()

df = yf.download(ticker, start = '2019-01-01', end = '2024-10-31')

df.head()
```

```python
df.info()
```

```python
plt.plot(df['Close'])

plt.plot(df['Open'])

plt.title(f"{ticker} Stock Graph")

plt.xlabel('Date')

plt.ylabel('Price (USD)')

plt.show()
```

```python
df = df.reset_index()
```

```python
df['Date'] = df['Date'].astype(str)
```

```python
df.info()
```

```python
df['Date'] = df['Date'].str[0:10]

df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df['ds'] = df['Date']
```

```python
df['y'] = df['Close']
```

```python
df.head()
```

```python
df = df[['ds', 'y']]
```

```python
df.head()
```

```python
from prophet import Prophet

model = Prophet()

model.fit(df)
```

```python
future = model.make_future_dataframe(periods = 1825)

future.tail()
```

```python
forecast = model.predict(future)

forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)

plt.title(f"{ticker} Stock Graph")

plt.xlabel('Date')

plt.ylabel('Closing Price (USD)')

plt.show()
```

```python
fig2 = model.plot_components(forecast)
```
